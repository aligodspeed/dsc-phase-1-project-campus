# Step 1: Importing the required libraries.

This step involves just importing the required libraries which are pandas, numpy. These are the necessary libraries when we need to work with data in data science.

```python
#Importing the necessary libraries.
import pandas as pd
import numpy as np
```

# Step 2: Getting the data-set from a different source and displaying the data-set.


This step involves getting the data-set from a different source (IMDB) which we are provided in zippedData folder.

```python
# I used a variable name to creat data_frame to call the table later easily.
imdb_title = pd.read_csv('zippedData/imdb.title.basics.csv.gz')
imdb_title
```

```python
# I used a variable name to creat data_frame to call the table later easily.
imdb_rate = pd.read_csv('zippedData/imdb.title.ratings.csv.gz')
imdb_rate
```

# Step 3: Checking the types of data.

Here we check the type of our data to see what kind of data are we working with.

```python
imdb_rate.info()
```

```python
imdb_title.info()
```

# Step 4: Merging dataFrames( Tables ).

Since we have to tables, and we need to use both tables, we Merge these two tables.

```python
# using this line of code to merge two dataFrame( tables.)
# these two tables have one column in Common which is " tconst" and we merge tables based on this column to make these two tables relevant.
new_imdb = pd.merge(imdb_rate,
                    imdb_title, on='tconst')
new_imdb
```

```python
# getting inof of the new table one more time to check the type of data we have.
new_imdb.info()
```

# Step 5: Removing the unused or irrelevant columns.

Often, not all of categories  or all columns of data in a dataset are useful in our analysis. We can drop those irrelevant Columns.

```python
# first, I made variable with columns that I need to drop.
# then, use one line of code to drop those columns.
# at the end, call the table with info to see if I droped them or not.


to_drop = ['primary_title','original_title', 'start_year',]
new_imdb.drop(to_drop, inplace=True, axis=1)
new_imdb
new_imdb.info()
```

# Step 6 : Dropping the missing or null values.

As you can see in the line above, there are some differences in two tittles( genres and runtime_minutes). These missing values make our result inaccurate. 

```python
# see the number of missing values
new_imdb.isna().sum()
```

```python
# drop those values with this line of code.
# using info to double check with values.
new_imdb.dropna(subset=['genres', 'runtime_minutes'], inplace= True)
new_imdb.info()

```

```python
# recall the table to see the final table with all changes.
new_imdb
```

# Step 7: Adding new columns if manipulating data is needed.

In this step, we add a column which is multiplying of two columns averagerating	and numvotes. I will explain later in lines [15] to [19]what is the reason to add this column. 

```python
#Multiplying averagerating and numvotes.
#name this new column rating*numvotes.
#add the rating*numvotes to the table. 

new_imdb['rating*numvotes']=new_imdb['averagerating']*new_imdb['numvotes']
new_imdb
```

To do so, I use the first genre at our list( Drama ) as a sample to show the procedure, and so on for the reat of the genres. But, first I need to show the difference between using only average rating and average rating with number of votes. 

```python
# making new table with using Drama as sample to develop our result on all genres. 
# in this table, we use averagerating and numvotes to see the differences. 

new_imdb_drama = new_imdb[new_imdb['genres']=='Drama']
new_imdb_drama
```

```python
# finding the average rating of Drama by using only averagerating.

just_average_drama = new_imdb[new_imdb['genres']=='Drama']['averagerating'].mean()
just_average_drama
```

```python

# find the sum of the numvotes column.

sum_votes = new_imdb_drama['numvotes'].sum()
sum_votes
```

```python
# find the sum of rating*numvotes column.

sum_rate_nvots = new_imdb_drama['rating*numvotes'].sum()
sum_rate_nvots
```

Now it's time to show the difference between the two average rating and average with number of votes. To do this, I use the fraction between the sum of rating*numvotes into sum of votes.

```python
# getting the average with this line of code.

drama_average = sum_rate_nvots / sum_votes
drama_average
```

As you can see there is significant difference between the number from line [16] which is the average of average rating, and the number from line [19] which is the average of rating with number of votes. # Visualization 

in this section we use some libraries and method to make bars and charts to show the result of our analysis.

```python
# import necessary libraries 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# line of codes to make bar based on the result of our dataFrame 

plt.figure(figsize=(18,6))

# make barplot and sort bars in descending order
sns.barplot(x='Genres', y='VotesRating', data=new_df_genres, 
            order=new_df_genres.sort_values('VotesRating',ascending = False).Genres)

# set labels
plt.xlabel("Genres", size=15)
plt.ylabel("Number of rating and votes", size=15)
plt.title("Top 10 Genres Based on Number of Votes and Rating", size=18)
plt.tight_layout()
```

```python
# line of codes to make bar based on the result of our dataFrame 

plt.figure(figsize=(18,6))

# make barplot and sort bars in descending order
sns.barplot(x='Genres', y='OnlyRate', data=new_df_genres_onlyRate, 
            order=new_df_genres_onlyRate.sort_values('OnlyRate',ascending = False).Genres)

# set labels
plt.xlabel("Genres", size=15)
plt.ylabel("Number of Only Rating", size=15)
plt.title("Top 10 Genres Based on only Rating", size=18)
plt.tight_layout()
```

```python
# line of codes to make bar based on the result of our dataFrame

plt.figure(figsize=(15,6))

# make barplot and sort bars in descending order
sns.barplot(x='Genres', y='NumVotes', data=new_df_number_of_votes, 
            order=new_df_number_of_votes.sort_values('NumVotes',ascending = False).Genres)

# set labels
plt.xlabel("Genres", size=15)
plt.ylabel("Number of Votes", size=15)
plt.title("Top 10 Genres Based on Number of Votes", size=18)
plt.tight_layout()
```

```python
# line of codes to make bar based on the result of our dataFrame


plt.figure(figsize=(10,6))

# make barplot and sort bars in descending order
sns.barplot(x='Runtime', y='Number of Movie', data=new_df_runtime, 
            order=new_df_runtime.sort_values('Number of Movie',ascending = False).Runtime)

# set labels
plt.xlabel("Run Time in Minute ", size=15)
plt.ylabel("Number of Movies ", size=15)
plt.title("Top 10 Runing Time in Minute ", size=18)
plt.tight_layout()
```
